```python
import pandas as pd
```

```python
d=pd.read_csv('C:/Users/HP/Desktop/file1.csv')
print(d)
```

```python
l=[a for a in d['regid']]
print(l)
```

```python
d.index=l
print(d)
```

```python
students={}
for i in l:
    temp=dict(d.loc[i])
    students[i]=temp
```

```python
for i in students.keys():
    print(i,students[i])
```

```python
#avg
for i in students.values():
    i['avg']=(i['phy']+i['chem']+i['math']+i['bio'])/4


#rank
a={ x['avg'] for x in students.values() }

a=list(a)
a.sort(reverse=True)
for i in students.values():
    i['rank']=a.index(i['avg'])+1
```

```python
first=list(students.keys())[0]
d=pd.DataFrame(students[first],index=[first])
for i in students.keys():
    if i==list(students.keys())[0]:
        continue
    d1=pd.DataFrame(students[i],index=[i])
    d=pd.concat([d,d1])
print(d)
```
