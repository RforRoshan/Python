```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x ** 2
y
```

```python
plt.plot(x, y, color='b')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("My first graph")
```

```python
plt.subplot(2,2,1)
plt.plot(x, y, color='b')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("My first graph")
plt.subplot(2,2,2)
plt.plot(x, y, color='b')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("My first graph")
plt.subplot(2,2,3)
plt.plot(x, y, color='b')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("My first graph")
plt.subplot(2,2,4)
plt.plot(x, y, color='b')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("My first graph")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([1, 1,1, 1])
axes1.plot(x, y, color='r')
axes1.set_title('First Axes')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
```

```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
fig = plt.figure()
axes1 = fig.add_axes([4, 1,1, 1])
axes1.plot(x, y, color='r')
axes1.set_title('First Axes')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
```

```python
fig = plt.figure() # create the figure object

axes1 = fig.add_axes([0.2, 0, 0.8, 0.8])
axes1.plot(x, y, color='r')
axes1.set_title('First Axes')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')

axes2 = fig.add_axes([1.1, 0.4, 0.8, 0.8])
axes2.plot(y, x, color='b')
axes2.set_title('Second Axes')
axes2.set_xlabel('Y')
axes2.set_ylabel('X')
```
