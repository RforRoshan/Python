```python
import pandas as pd
```

```python
df=pd.read_csv("chipotle.tsv",sep="\t")
df.head()
```

```python
df["item_price"]=df["item_price"].apply(lambda s: float(s[1:]))
df.head()
```

```python
df.where((df["item_price"]/df["quantity"])>10).count()["item_price"]
```

```python
t=df[["item_name","item_price"]]
t
```

```python
t=t.sort_values(by=["item_name"])
t.reset_index(inplace = True)
t
```

```python
t1=t.where(t["item_price"]==t["item_price"].max()).dropna()
t1
```
